# Visualizing the `distribution` of a dataset

[Tutorial](http://seaborn.pydata.org/tutorial.html#tutorial)

## 1 - Univariate distribution

````python
%matplotlib inline
````

````python
import matplotlib as mpl
````

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats, integrate
````

````python
sns.set(color_codes=True)
````

````python
np.random.seed(sum(map(ord, "distributions")))
````

````python
mean, cov =[0,1], [(1,.5), (0.5,1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
````

### [`sns.distplot()`](http://seaborn.pydata.org/generated/seaborn.distplot.html#seaborn.distplot)
```python
sns.distplot(a, bins=None, hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, color=None, vertical=False, norm_hist=False, axlabel=None, label=None, ax=None)
```
plot distribution
- `hist`: 是否显示直方图
- `kde`: 是否显示密度曲线
- `rug`: 是否显示“小地毯”

````python
x=np.random.normal(size=100)
sns.distplot(x,rug=True,);
````

````python
sns.distplot(x, bins=30, rug=True, hist=False);
````

### [`sns.kdeplot()`](http://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot)
`sns.kdeplot(data, data2=None, shade=False, vertical=False, kernel='gau', bw='scott', gridsize=100, cut=3, clip=None, legend=True, cumulative=False, shade_lowest=True, cbar=False, cbar_ax=None, cbar_kws=None, ax=None, **kwargs)`
- `cut` : 曲线两端延伸长度，只影响曲线的位置和比例 ，不影响拟合的结果
- `bw` : 估算概率密度时，直方图的 `bandwidth`

````python
sns.kdeplot(x, label="guess")
sns.kdeplot(x, bw=.2, label="bw: 0.2",lw=3)
sns.kdeplot(x, bw=2, label="bw: 2");
````

````python
plt.subplot(141)
sns.kdeplot(x, bw=2, label="cut: 0",cut=0);
plt.subplot(142)
sns.kdeplot(x, bw=2, label="bw: 6",cut=6);
plt.subplot(143)
sns.kdeplot(x, bw=2, cut=1 ,label="cut: 1");
plt.subplot(144)
sns.kdeplot(x, bw=2);
plt.tight_layout()
````

## 2 - Bivariate distributions

### [`sns.jointplot()`](http://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot)
`sns.jointplot(x, y, data=None, kind='scatter', stat_func=<function pearsonr>, color=None, size=6, ratio=5, space=0.2, dropna=True, xlim=None, ylim=None, joint_kws=None, marginal_kws=None, annot_kws=None, **kwargs)`
- `kind` : scatter, kde, hex

````python
plt.scatter(df["x"],df["y"]);
````

````python
sns.jointplot(x="x", y="y", data=df,xlim=(-3,4));
````

````python
with sns.axes_style("white"):
    sns.jointplot(x="x", y="y", data=df, kind="hex", color="k")
````

`jointplot()` returns the `JointGrid` object after plotting

#### [`plt.scatter()`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter)
`plt.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, hold=None, data=None)`
- [`matplotlib.markers`](http://matplotlib.org/api/markers_api.html#module-matplotlib.markers)

````python
g = sns.JointGrid(df.x, df.y, ylim=(-1,4),size=5, space=0.1)
g = g.plot_joint(plt.scatter, c="g", s=30, linewidth=1, marker="+")
g = g.plot_marginals(sns.kdeplot, shade=True)
g.ax_joint.collections[0].set_alpha(0.5)
g.set_axis_labels("$X$", "$Y$");
````

### `sns.rugplot()`

````python
f, axes= plt.subplots(ncols=2,figsize=(16,5))
sns.kdeplot(data=df,ax=axes[0]);
sns.rugplot(df.x, ax=axes[0]);
sns.rugplot(df.y, vertical=True,ax=axes[0]);
sns.distplot(df.x,ax=axes[1]);
````

### `cmap`

````python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.x, df.y, cmap=cmap,shade=True, n_levels=60);
````

### [`sns.JiontGrid()`](http://seaborn.pydata.org/generated/seaborn.JointGrid.html#seaborn.JointGrid)
`sns.JointGrid(x, y, data=None, size=6, ratio=5, space=0.2, dropna=True, xlim=(0,5), ylim=(1,3))`
- `ratio` : Ratio of joint axes size to marginal axes height.
- `g.ax_marg_x` : AxesSubplot 对象，`matplotlib` 的子绘图区域
- `g.ax_marg_y` : 同上

````python
g = sns.JointGrid(x="x", y="y", data=df)
g.plot(plt.scatter, sns.kdeplot);
````

````python
g = sns.JointGrid(x="x", y="y", data=df)
g = g.plot_joint(sns.regplot, color="g")
g = g.plot_marginals(sns.distplot, color="r")
````

````python
g = sns.JointGrid(df.x, df.y, space=0.1, size=5)
_ = g.plot_joint(sns.regplot, color="g")
sns.kdeplot(df.x, shade=True, ax=g.ax_marg_x)
_ = g.ax_marg_y.hist(df.y, orientation="horizontal",color="r")
````

`sns.JointGrid(..., ration=2)`

`sns.JointGrid(..., space=1, size=5, ration=2)`


````python
g = sns.JointGrid(x="x", y="y", data=df, size=5, ratio=6, ylim=(-1.5,4))
g = g.plot_joint(plt.scatter, color="g", edgecolor="white")
g = g.plot_marginals(sns.kdeplot)
g = g.annotate(stats.pearsonr)
````

## 3 - Multiple pairwise bivarite dist

### [`sns.pairplot()`](http://seaborn.pydata.org/generated/seaborn.pairplot.html#seaborn.pairplot)
```python
sns.pairplot(data, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='hist', markers=None, size=2.5, aspect=1, dropna=True, plot_kws=None, diag_kws=None, grid_kws=None)
```
- `x_i, x_j` 两两绘图
- `i=j` 时， `"dist"`
- `i≠j` 时， `"scatter"` or `"reg"`
- `kind` : `"scatter"` or `"reg"`

````python
iris = sns.load_dataset("iris")
sns.pairplot(iris, markers="+", kind="reg");
````

````python
sns.pairplot(df);
````

### [`sns.PairGrid()`](http://seaborn.pydata.org/generated/seaborn.PairGrid.html#seaborn.PairGrid) 
```python
sns.PairGrid(data, hue=None, hue_order=None, palette=None, hue_kws=None, vars=None, x_vars=None, y_vars=None, diag_sharey=True, size=2.5, aspect=1, despine=True, dropna=True)
```
- `g.map_diag(...)`
- `g.map_offdiag(...)`

````python
g = sns.PairGrid(iris)

#绘制对角线区域：i=j
g.map_diag(sns.kdeplot);  

#绘制非对角线区域：i≠j
g.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6);
````
